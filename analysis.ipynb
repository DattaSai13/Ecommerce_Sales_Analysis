```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("sales_data.csv")
df["Date"] = pd.to_datetime(df["Date"])
df.head()
```

```python
df.info()
df.describe()
```

```python
df.fillna(0, inplace=True)
df.drop_duplicates(inplace=True)
```

```python
df.groupby("Product")["Total_Sales"].sum()
```

```python
df.groupby("Region")["Total_Sales"].sum()
```

```python
df.groupby(df["Date"].dt.month)["Total_Sales"].sum()
```

```python
df.groupby("Product")["Total_Sales"].sum().plot(kind="bar")
plt.title("Sales by Product")
plt.show()
```

```python
df.groupby(df["Date"].dt.month)["Total_Sales"].sum().plot(kind="line", marker="o")
plt.title("Monthly Sales Trend")
plt.show()
```

```python
df.groupby("Region")["Total_Sales"].sum().plot(kind="pie", autopct="%1.1f%%")
plt.title("Sales by Region")
plt.ylabel("")
plt.show()
```
